```python
import pandas as pd

by_url=pd.read_csv("HM Raw Data Binning.csv")
# Sort by url, then by ascending position, then largest search volume
by_url = by_url.sort_values(['url', 'position', 'search_volume'],
              ascending = [True, True, False])
# Only have highest position if keyword ranks multiple times on same URL
by_url = by_url.drop_duplicates(subset=['url', 'keyword'], keep='first')
```

```python
by_url.head()
```

```python
# List of all URLs
url_list = by_url.url.unique()

condenced_df = pd.DataFrame()


for url in url_list:
    rslt_df = pd.DataFrame()
    # Filter to only URL
    rslt_df = by_url.loc[by_url['url'] == url]
    # Get top 10
    rslt_df = rslt_df.iloc[:10]
    # Calculate needed values
    url_count = int(len(rslt_df.index))
    samples = "\n".join(rslt_df['keyword'])
    min_rank = rslt_df['position'].min()
    max_rank = rslt_df['position'].max()
    # Avg position, Sum of search volume, and Sum of number of results
    # TODO: Add in monthly search volume sum
    rslt_df = rslt_df.groupby(["url"], as_index=False).agg(
        {"position": "median", "search_volume": "sum", "number_of_results": "sum"})
    # Add to result df
    rslt_df["min_rank"] = min_rank
    rslt_df["max_rank"] = max_rank
    rslt_df["samples"] = samples
    rslt_df["keyword_count"] = url_count
    # Append to overall df
    condenced_df = condenced_df.append(rslt_df)
```
